What is machine learning?
A study that gives computers the ability to learn without being explicitly programmed. (rthur Samuel 1959)

There are 2 types of machine learning algorithms.
1. Supervised learning. 
2. Unsupervised Learning

# Supervised learning.
A model is trained with inputs(X) and outputs (Y) data to predict only outputs later.
input/Feature(x) → output/Target(y)

There are 2 Types
1. Regression 
2. Classification

# Regression 
Predicting a number from infinitely many possible numbers. 
For example, home price prediction based on sizes.

# Classification
Predict small finite set of outputs such as 0, 1 and not in between. Classification predicts categories. 
For example, breast cancer detection where 0 = malignant, 1 = benign based on tumor sizes.
(Tumor Size, Diagnosis): (2, 0), (5, 1), (1,0), (7, 1)
Possible outcome is either zero or one and can also be plotted as a number line.

# Unsupervised learning.
Algorithm finds patterns in unlabeled data basically. 
There are three Types
1. Clustering - 
Takes data without labels and tries to grouop them. For exaample DNA , Google search suggestions

2. Anomaly Detection
Algorithm finds unusual data points

3. Dimensionality Reduction 
Compresses data using fewer numbers

# REGRESSION TERMINOLOGIES
1. Training Set - Data used to train the model
2. x - Input / Feature variable
3. y - Output / Target variable
4. m - total number of training examples
5. (x, y) - A single training example
6. (x(i), y(i)) - ith training example

# DESIGN FLOW
        Training Set{Feature, Target}
                  ↓
        Learning Algorithm
                  ↓
            x(feature) → Function(model) → ŷ(prediction)
                  y = target
                  ŷ = prediction

# REPRESENTING F
Fw,b(x) = wx + b
F(x) = wx + b
where w = weight, b = bias

# Note 
Linear regression with one variable is referred to as Univariate Linear Regression

# Notations and Representation in Python
a             → Scalar  
**a** (bold)  → Vector
x             → Feature      →  x_train
y             → Target       →  y_train
x(i), y(i)                   →  x_i , y_i
Fw,b(x(i))                   →  F_wb 
m as m
w as w
b as b

would like to fit a linear regression model through these two points, so you can then predict price for other houses - say, a house with 1200 sqft.

```python
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')  #style plot is in the course cloud hence cant run locally
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

numpy shape parameter(.shape) will be used to return a tuple with an entry for each dimension. .shape[0] is the length pf the array and number of examples(m)

```python
#NUMBER OF TRAINING EXAMPLES
print(f"x_train shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples : {m}")
m = len(x_train)
print(f"Number of training examples : {m}")
```

```python
#TRAINING EXAMPLES 
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^(0), y^(0)) = ({x_i}, {y_i})")

i = 1
```

```python
i = 1
x_i = x_train[i]
y_i = y_train[i]
print (f"(x^(1), (y^(1)) = ({x_i}, {y_i})")
```

```python
#PLOTTING THE DATA
plt.scatter(x_train, y_train, marker='x', c ='red')
plt.title("Housing Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (1000s of $)")
plt.show()
```

 # Visualization Syntax

 plt.plot: Used to visualize the linear regression as a line 
 Labeling: Always include plt.xlabel, plt.ylabel, and plt.title 

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")

f_wb = w * x_i + b
print(f"f_wb: {f_wb}")
```

```python
def compute_model_output (x, w, b):
    m = x_train.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b 
    return f_wb
tmp_f_wb = compute_model_output(x_train, w, b)
print (f"temp f: {tmp_f_wb}")
```

```python
#PLOTTING MODEL PREDICTION
plt.plot(x_train, tmp_f_wb, c='b', label='Prediction')

#PLOTTING THE DATA
plt.scatter(x_train, y_train, marker='x', c ='red', label='Training Data')

plt.title("Housing Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (1000s of $)")
plt.legend()
plt.show()
```
