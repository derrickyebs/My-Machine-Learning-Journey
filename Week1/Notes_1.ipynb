What is machine learning?
A study that gives computers the ability to learn without being explicitly programmed. (rthur Samuel 1959)

# Two types of machine learning algorithms.
1. Supervised learning. 
2. Unsupervised Learning

# Supervised learning.
A model is trained with inputs(X) and outputs (Y) data to predict only outputs later.
input/Feature(x) → output/Target(y)

# Two Types
1. Regression 
2. Classification

# Regression 
Predicting a number from infinitely many possible numbers. 
For example, home price prediction based on sizes.

# Classification
Predict small finite set of outputs such as 0, 1 and not in between. Classification predicts categories. 
For example, breast cancer detection where 0 = malignant, 1 = benign based on tumor sizes.
(Tumor Size, Diagnosis): (2, 0), (5, 1), (1,0), (7, 1)
Possible outcome is either zero or one and can also be plotted as a number line.

# Unsupervised learning.
Algorithm finds patterns in unlabeled data basically
There are three Types
1. Clustering - 
Takes data without labels and tries to grouop them. For exaample DNA , Google search suggestions

2. Anomaly Detection
Algorithm finds unusual data points

3. Dimensionality Reduction 
Compresses data using fewer numbers

# REGRESSION TERMINOLOGIES
1. Training Set - Data used to train the model
2. x - Input / Feature variable
3. y - Output / Target variable
4. m - total number of training examples
5. (x, y) - A single training example
6. (x(i), y(i)) - ith training example

# DESIGN FLOW
        Training Set{Feature, Target}
                  ↓
        Learning Algorithm
                  ↓
            x(feature) → Function(model) → ŷ(prediction)
                  y = target
                  ŷ = prediction

# REPRESENTING F
Fw,b(x) = wx + b
F(x) = wx + b
where w = weight, b = bias

# Note 
Linear regression with one variable is referred to as Univariate Linear Regression

# Notations and Representation in Python
a             → Scalar  
**a** (bold)  → Vector
x             → Feature      →  x_train
y             → Target       →  y_train
x(i), y(i)                   →  x_i , y_i
Fw,b(x(i))                   →  F_wb 
m as m
w as w
b as b
