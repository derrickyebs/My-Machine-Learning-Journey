What is machine learning?
A study that gives computers the ability to learn without being explicitly programmed. (rthur Samuel 1959)

There are 2 types of machine learning algorithms.
1. Supervised learning. 
2. Unsupervised Learning

# Supervised learning.
A model is trained with inputs(X) and outputs (Y) data to predict only outputs later.
input/Feature(x) → output/Target(y)

There are 2 Types
1. Regression 
2. Classification

# Regression 
Predicting a number from infinitely many possible numbers. 
For example, home price prediction based on sizes.

# Classification
Predict small finite set of outputs such as 0, 1 and not in between. Classification predicts categories. 
For example, breast cancer detection where 0 = malignant, 1 = benign based on tumor sizes.
(Tumor Size, Diagnosis): (2, 0), (5, 1), (1,0), (7, 1)
Possible outcome is either zero or one and can also be plotted as a number line.

# Unsupervised learning.
Algorithm finds patterns in unlabeled data basically. 
There are three Types
1. Clustering - 
Takes data without labels and tries to grouop them. For exaample DNA , Google search suggestions

2. Anomaly Detection
Algorithm finds unusual data points

3. Dimensionality Reduction 
Compresses data using fewer numbers

# REGRESSION TERMINOLOGIES
1. Training Set - Data used to train the model
2. x - Input / Feature variable
3. y - Output / Target variable
4. m - total number of training examples
5. (x, y) - A single training example
6. (x(i), y(i)) - ith training example

# DESIGN FLOW
        Training Set{Feature, Target}
                  ↓
        Learning Algorithm
                  ↓
            x(feature) → Function(model) → ŷ(prediction)
                  y = target
                  ŷ = prediction

# Cost Function Formula
Tells how well a model is doing and to try to make it do better
w, b are the parameters. They are variables you adjust during training to improve the model.

Goal is to minimize the cost function represented as J(w,b) which is the average of the sum of the difference in prediction and target

# Gradient Descent 
Systematic way to find the values of w  and b to find the smallest possible cost function J(w,b). Not only for linear regression

Goal: Minimize J(w,b)
Outline: Start with some w,b say (w=0, b=0), Keep changing w and b to reduce J(w,b) until we setele at or near a minimum
Note: For functions that may not be a soup bowl, there may be more than one minimum.
Linear Regressions with squared error cost function alwasy have a bowl shape

Gradient Descent(GD) helps you go downhill in the hill, valley analogy as given by the author. The multiple minimum valley points are called Local Minima

# Gradient Descent Algorithm
w = w - alpha times partial derivative of J(w,b) with respect to(w.r.t) w
b = b - alpha times partial derivative of J(w,b) w.r.t. to b

where alpha =  learning rate
